```python
# The imports

from dotenv import load_dotenv
from agents import Agent, Runner, trace, add_trace_processor
from agents.tracing.processors import ConsoleSpanExporter, BatchTraceProcessor
from azure_client import get_AzureOpenAIChatCompletionsModel
```

```python
# The usual starting point

load_dotenv()

console_exporter = ConsoleSpanExporter()
console_processor = BatchTraceProcessor(console_exporter)
add_trace_processor(console_processor)
```

```python

model = get_AzureOpenAIChatCompletionsModel()

# Make an agent with name, instructions, model
agent = Agent(name="Jokester", instructions="You are a joke teller", model=model)
```

```python
# Run the joke with Runner.run(agent, prompt) then print final_output

with trace("Telling a joke"):
    result = await Runner.run(agent, "Tell a joke about Autonomous AI Agents")
    print(result.final_output)
```
