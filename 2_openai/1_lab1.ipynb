```python
# The imports

from dotenv import load_dotenv
from agents import Agent, Runner, trace, add_trace_processor, OpenAIChatCompletionsModel
from openai import AsyncAzureOpenAI
import os
from agents.tracing.processors import ConsoleSpanExporter, BatchTraceProcessor

```

```python
# The usual starting point

load_dotenv(override=True)

# Azure OpenAI configuration (set these in your .env)
# AZURE_OPENAI_API_KEY, AZURE_OPENAI_ENDPOINT, AZURE_OPENAI_API_VERSION
# AZURE_OPENAI_CHAT_DEPLOYMENT_NAME (or AZURE_OPENAI_DEPLOYMENT)

azure_client = AsyncAzureOpenAI(
    api_key=os.getenv("AZURE_OPENAI_API_KEY"),
    api_version=os.getenv("AZURE_OPENAI_API_VERSION") or "2024-12-01-preview",
    azure_endpoint=os.getenv("AZURE_OPENAI_ENDPOINT"),
)

deployment_name = os.getenv("AZURE_OPENAI_CHAT_DEPLOYMENT_NAME") or os.getenv("AZURE_OPENAI_DEPLOYMENT")

if not deployment_name:
    raise ValueError("Set AZURE_OPENAI_CHAT_DEPLOYMENT_NAME (or AZURE_OPENAI_DEPLOYMENT) in your .env")


console_exporter = ConsoleSpanExporter()
console_processor = BatchTraceProcessor(console_exporter)
add_trace_processor(console_processor)
```

```python

# Make an agent with name, instructions, model
# For Azure, `model` should be your deployment name.

agent = Agent(name="Jokester", instructions="You are a joke teller", model=OpenAIChatCompletionsModel(openai_client=azure_client, model=deployment_name))
```

```python
# Run the joke with Runner.run(agent, prompt) then print final_output

with trace("Telling a joke"):
    result = await Runner.run(agent, "Tell a joke about Autonomous AI Agents")
    print(result.final_output)
```
