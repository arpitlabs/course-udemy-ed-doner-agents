### Wait, did that just output `False`??

If so, the most common reason is that you didn't save your `.env` file after adding your keys! Be sure to have saved.

Also, make sure the `.env` file is named precisely `.env` and is in the project root directory (`agents`). This lab expects **Azure OpenAI** variables (`AZURE_OPENAI_API_KEY`, `AZURE_OPENAI_ENDPOINT`, `AZURE_OPENAI_API_VERSION`, and `AZURE_OPENAI_CHAT_DEPLOYMENT_NAME` or `AZURE_OPENAI_DEPLOYMENT`).

By the way, your `.env` file should have a stop symbol next to it in Cursor on the left, and that's actually a good thing: that's Cursor saying to you, "hey, I realize this is a file filled with secret information, and I'm not going to send it to an external AI to suggest changes, because your keys should not be shown to anyone else."

```python
from utils import azure_client, MODEL
```

```python
# Create a list of messages in the familiar OpenAI format

messages = [{"role": "user", "content": "What is 2+2?"}]
```

```python
# Call your deployed model on Azure (`CHAT_DEPLOYMENT` in the cell above).

response = azure_client.chat.completions.create(
    model=MODEL,
    messages=messages,
)

print(response.choices[0].message.content)
```

```python
# And now - let's ask for a question:

question = "Please propose a hard, challenging question to assess someone's IQ. Respond only with the question."
messages = [{"role": "user", "content": question}]
```

```python
# Same deployment as above (Azure maps one deployment name to one model in your resource).

response = azure_client.chat.completions.create(
    model=MODEL,
    messages=messages,
)

question = response.choices[0].message.content

print(question)
```

```python
response = azure_client.chat.completions.create(
    model=MODEL,
    messages=messages,
)

answer = response.choices[0].message.content
print(answer)
```

            Then ask the LLM to present a pain-point in that industry - something challenging that might be ripe for an Agentic solution.<br/>

```python
# First create the messages:

messages = [{"role": "user", "content": "Pick a business area that might be worth exploring for an Agentic AI opportunity."}]

# Then make the first call:

response = azure_client.chat.completions.create(
    model=MODEL,
    messages=messages,
)

# Then read the business idea:
business_idea = response.choices[0].message.content
messages = [{"role": "user", "content": "Pick a pain point in the business area: '" + business_idea + "'"}]

response = azure_client.chat.completions.create(
    model=MODEL,
    messages=messages,
)

# finally, propose an agentic AI solution
pain_point = response.choices[0].message.content
messages = [{"role": "user", "content": "Propose an Agentic AI solution for the pain point: '" + pain_point + "'"}]

response = azure_client.chat.completions.create(
    model=MODEL,
    messages=messages,
)

agentic_solution = response.choices[0].message.content

# display the agentic solution
from IPython.display import Markdown, display
display(Markdown(agentic_solution))
```
